```python
tasks = []  # This list will store all the tasks.


# --- Function to show all tasks ---
def show_tasks():
    if len(tasks) == 0:
        print("No tasks added yet ðŸ˜Š")
    else:
        print("\nyour task:")
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")
    print()

# --- Main loop ---
while True:
    print("1. Add Task")
    print("2. Delete Task")
    print("3. Show Tasks")
    print("4. Exit")

    choice = input("what are you going to do? (1-4): ")

    if choice == "1":
        task = input("Which task do you want to do? ")
        tasks.append(task)
        print("task added! âœ”\n")

    elif choice == "2":
        show_tasks()
        if len(tasks) == 0:
            continue
        num = int(input("Which task should I delete? Tell me the number.: "))
        if 1 <= num <= len(tasks):
            tasks.pop(num - 1)
            print("Task deleted! âŒ\n")
        else:
            print("gave wrong numberðŸ˜…\n")

    elif choice == "3":
        show_tasks()

    elif choice == "4":
        print("thanks bye! ðŸ‘‹")
        break

    else:
        print("choose from 1-4 any numberÂ ðŸ™‚\n")
```
